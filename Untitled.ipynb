```python

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('lan_img1.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
plt.figure(figsize = (20, 10))
plt.subplot(1,2,1)
plt.imshow(img)
plt.title('Input Image')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(gray, cmap='gray')
plt.title('Grayscale')
plt.axis('off')
```

```python
threshold = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)[1]
```

```python
plt.figure(figsize = (20, 10))
plt.subplot(1,1,1); plt.imshow(threshold, cmap = 'gray'); plt.title('Threshold');

plt.imshow(threshold, cmap='gray')
plt.title('Threshold')
plt.axis('off')
```

```python

roi_vertices = np.array([[[100, 540],
                          [900, 540],
                          [515, 320],
                          [450, 320]]])

mask = np.zeros_like(threshold)   
if len(threshold.shape) > 2:
    channel_count = threshold.shape[2]  # 3 or 4 depending on the image.
    ignore_mask_color = (255,) * channel_count
else:
    ignore_mask_color = 255


cv2.fillPoly(mask, roi_vertices, ignore_mask_color)
roi = cv2.bitwise_and(threshold, mask)

plt.figure(figsize = (20, 10))
plt.subplot(1,3,1); plt.imshow(threshold, cmap = 'gray'); plt.title('Initial threshold')
plt.subplot(1,3,2); plt.imshow(mask, cmap = 'gray');      plt.title('Polyfill mask')
plt.subplot(1,3,3); plt.imshow(roi, cmap = 'gray');       plt.title('Isolated roi');
```

```python
edges = cv2.Canny(gray, 50, 150)
```

```python
canny_blur = cv2.GaussianBlur(edges, (5, 5), 0)
```

```python

plt.figure(figsize = (20, 10))
plt.subplot(1,2,1); plt.imshow(edges, cmap = 'gray'); plt.title('Edge detection')
plt.subplot(1,2,2); plt.imshow(canny_blur, cmap = 'gray'); plt.title('Blurred edges');
plt.subplot(1,2,1)
plt.imshow(edges, cmap='gray')
plt.title('Edge Detection')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(canny_blur, cmap='gray')
plt.title('Blurred Image')
plt.axis('off')
```

```python
def draw_lines(img, lines, color = [255, 0, 0], thickness = 2):
    """Utility for drawing lines."""
    if lines is not None:
        for line in lines:
            for x1,y1,x2,y2 in line:
                cv2.line(img, (x1, y1), (x2, y2), color, thickness)
```

```python
lines = cv2.HoughLinesP(
    canny_blur,
    rho=1,
    theta=np.pi/180,
    threshold=50,
    minLineLength=50,
    maxLineGap=100
)
```

```python
def average(values):
    n = len(values)
    return sum(values) / n
```

```python
def extrapolate_lines(lines):
    slopes = []
    consts = []

    for line in lines:
        x1, y1, x2, y2 = line[0]
        
        if x2 - x1 != 0:  
            slope = (y2 - y1) / (x2 - x1)
            c = y1 - slope * x1

            slopes.append(slope)
            consts.append(c)

    return slopes, consts
```
